```python
timeperiods =['2000-01-01','2004-01-01','2008-01-01','2012-01-01','2016-01-01','2020-01-01','2025-07-17']
for i in range(0,len(timeperiods)-1):
    snapshot_aapl=aapl_historical[(aapl_historical.index > timeperiods[i]) & (aapl_historical.index < timeperiods[i+1])]
    snapshot_sandp500=sandp500_historical[(sandp500_historical.index > timeperiods[i]) & (sandp500_historical.index < timeperiods[i+1])] 

    fig=plt.figure(figsize=(15,5))
    ax=plt.axes()
    ax.plot(snapshot_aapl.index,snapshot_aapl["Close_Rescaled"])
    ax.plot(snapshot_sandp500.index,snapshot_sandp500["Close_Rescaled"])
    
    plt.xlabel("Date")
    plt.ylabel("Close")
    plt.legend(["apple","s and p 500"])
    
```

```python

snapshot_aapl_2000_2004=aapl_historical[(aapl_historical.index > '2007-01-01') & (aapl_historical.index < '2014-01-01')]
snapshot_sandp500_2000_2004=sandp500_historical[(sandp500_historical.index > '2007-01-01') & (sandp500_historical.index < '2014-01-01')] 

fig=plt.figure(figsize=(15,5))
ax=plt.axes()
ax.plot(snapshot_aapl_2000_2004.index,snapshot_aapl_2000_2004["Close_Rescaled"])
ax.plot(snapshot_sandp500_2000_2004.index,snapshot_sandp500_2000_2004["Close_Rescaled"])

ax.annotate('Introduction of Iphone', xy=(pd.to_datetime("2007-01-09"),330), xytext=(pd.to_datetime("2007-01-09"),600),ha="center",
arrowprops=dict(facecolor='black' ,shrink=0.005))

ax.annotate('Global Financial Crisis', xy=(pd.to_datetime("2008-01-14"),638.85), xytext=(pd.to_datetime("2008-01-14"),900),ha="center",
arrowprops=dict(facecolor='black' ,shrink=0.005))

ax.annotate('Death of steve jobs', xy=(pd.to_datetime("2011-05-10"),1248.732485), xytext=(pd.to_datetime("2011-05-10"),1500),ha="center",
arrowprops=dict(facecolor='black' ,shrink=0.005))

ax.annotate('Iphone 5 launched', xy=(pd.to_datetime("2012-09-21"),2512.45), xytext=(pd.to_datetime("2012-09-21"),2700),ha="center",
arrowprops=dict(facecolor='black' ,shrink=0.005))

ax.annotate('Iphone 5c launched', xy=(pd.to_datetime("2013-09-10"),1817.4), xytext=(pd.to_datetime("2013-09-10"),2000),ha="center",
arrowprops=dict(facecolor='black' ,shrink=0.005))
    
plt.xlabel("Date")
plt.ylabel("Close")
plt.legend(["apple","s and p 500"])
    
```
